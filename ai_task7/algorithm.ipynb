````python
class Graph:
    def __init__(self, adjacency_list):
        self.adjacency_list = adjacency_list

    def get_neighbors(self, v):
        return self.adjacency_list[v]

    def h(self, n):
        H = {'A': 1, 'B': 1, 'C': 1, 'D': 1}
        return H[n]

    def a_star_algorithm(self, start_node, stop_node):
        open_list = set([start_node])
        closed_list = set([])
        g = {start_node: 0}
        parents = {start_node: start_node}

        while len(open_list) > 0:
            n = None
            for v in open_list:
                if n is None or g[v] + self.h(v) < g[n] + self.h(n):
                    n = v
            if n is None:
                print('Path does not exist!')
                return None
            if n == stop_node:
                path = []
                while parents[n] != n:
                    path.append(n)
                    n = parents[n]
                path.append(start_node)
                path.reverse()
                print('Path found:', path)
                return path
            for (m, weight) in self.get_neighbors(n):
                if m not in open_list and m not in closed_list:
                    open_list.add(m)
                    parents[m] = n
                    g[m] = g[n] + weight
                else:
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight
                        parents[m] = n
                        if m in closed_list:
                            closed_list.remove(m)
                            open_list.add(m)
            open_list.remove(n)
            closed_list.add(n)
        print('Path does not exist!')
        return None

adjacency_list = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('D', 5)],
    'C': [('D', 12)]
}
graph1 = Graph(adjacency_list)
graph1.a_star_algorithm('A', 'D')
````

````python
# Min-Max Algorithm / Game Playing (AO* Algorithm)
# odd = max , even = min
import math
values = [3,2,5,4,7,2,7,7]
levels = math.log(len(values), 2) # log of 8 with base 2

def minmax(currDepth, nodeIndex , minTurn, values, treeDepth):
        if currDepth == treeDepth:  #agar limit/leaves reach hogaye h /last pe pohonch gaye
            return values[nodeIndex]

        if(minTurn):  # agar minturn h to ye func chle ga
            return min(minmax(currDepth+1, nodeIndex*2 , False, values, treeDepth), 
                       minmax(currDepth+1, nodeIndex*2+1 , False, values, treeDepth))
        
        else: # wrna ye maxturn func chle ga
             return max(minmax(currDepth+1, nodeIndex*2 , True, values, treeDepth),
                        minmax(currDepth+1, nodeIndex*2+1 , True, values, treeDepth))


print(minmax(0, 0, False, values, levels))
# zero lvl se start hoga 
# 0 ka matlb left, 1 ka matlab right.
# min agar else wale pe lagaye ge , to AO* Algorithm chle ga , wrna ye min-max game chl rahi h.

````

## CODE LOGIC 


### 1) `minTurn` (True/False) â€” iska matlab

* `minTurn = True` â†’ **Minimizer** ki turn hai: is node pe opponent ki strategy hai, to woh *chhota* value choose karega (use `min()`).
* `minTurn = False` â†’ **Maximizer** ki turn hai: is node pe hum/agent ka turn hai, to woh *bara* value choose karega (use `max()`).

Agar root pe `minTurn=True` pass kia hai to root se start mein minimizer move karega. Agar hum chahen root se maximizer start kare to `minTurn=False` pass karo.

---

### 2) `levels = int(math.log(len(values), 2))`

* Agar leaves 8 hain â†’ `log2(8) = 3` â†’ toh tree depth = 3.
* Depth counting: root = depth 0, leaves = depth 3.
  Base case: jab `currDepth == treeDepth` â†’ leaf reached â†’ `return values[nodeIndex]`.

---

### 3) `nodeIndex`, `nodeIndex*2`, `nodeIndex*2 + 1` ka matlab (array-as-tree mapping)

Yeh **binary tree stored level-by-level** ka convention hai. Har depth pe nodes ka index `0..(2^depth -1)` hota hai.

* Root: `currDepth=0`, `nodeIndex=0`.
* Uska left child (next depth) â†’ `nodeIndex_left = nodeIndex * 2`.
* Uska right child â†’ `nodeIndex_right = nodeIndex * 2 + 1`.

Why? Kyunki agar aap kisi depth `d` pe node ka position `i` maan lo, to next depth (d+1) pe us node ke children sequentially positions `2*i` aur `2*i+1` lete hain. Yeh *binary heap / complete binary tree* style indexing hai.

**Example mapping (leaves only)**

* Depth 3 (leaves) ke indices: `0,1,2,3,4,5,6,7` â†’ ye directly `values` array ke indices hain.

### 4) Ab **step-by-step trace** tumhare `values = [3,2,5,4,7,2,7,7]` aur `minTurn=True` (root is Min)

Tree depth = 3. Start: `minmax(0, 0, True, values, 3)`

Main important recursive calls aur returns niche diye hain:

```
Call A: minmax(depth=0, index=0, minTurn=True)   # root (Min)
    -> will return min( LeftResult, RightResult )

 Left branch of root:
 Call B: minmax(depth=1, index=0, minTurn=False)  # Max
    -> returns max( B_left, B_right )

    B_left:
    Call C: minmax(depth=2, index=0, minTurn=True)  # Min
        -> returns min( values[0], values[1] ) = min(3,2) = 2

    B_right:
    Call D: minmax(depth=2, index=1, minTurn=True)  # Min
        -> returns min( values[2], values[3] ) = min(5,4) = 4

    So Call B returns max(2, 4) = 4

 Right branch of root:
 Call E: minmax(depth=1, index=1, minTurn=False)  # Max
    -> returns max( E_left, E_right )

    E_left:
    Call F: minmax(depth=2, index=2, minTurn=True)  # Min

    E_right:

    So Call E returns max(2, 7) = 7

Finally root (Call A) returns min( LeftResult=4, RightResult=7 ) = 4
```

### 6) Kyu aise implement kiya? (logic behind design)

* Humne tree ko **implicit array-style** represent kiya â€” sirf leaves (final scores) diye gaye hain.
* `nodeIndex` har depth par batata hai *us node ka position* us depth mein.
* `nodeIndex*2` aur `nodeIndex*2+1` se hum left/right child pe ja rahe hain.
* Recursion se hum **bottom-up** values compute karte hain: leaf values â†’ parents combine (min or max) â†’ next parent... â†’ root result.
* `minTurn` boolean decide karta hai kaunsa function (min ya max) lagaana hai at that node.

import math
values = [3,2,5,4,7,2,7,7]

def minimax(currDepth, nodeIndex, isMinTurn, values, treeDepth):
    if currDepth == treeDepth:
        return values[nodeIndex]

    left = minimax(currDepth+1, nodeIndex*2,     not isMinTurn, values, treeDepth)
    right= minimax(currDepth+1, nodeIndex*2 + 1, not isMinTurn, values, treeDepth)

    return min(left, right) if isMinTurn else max(left, right)

print(minimax(0, 0, True, values, treeDepth))
