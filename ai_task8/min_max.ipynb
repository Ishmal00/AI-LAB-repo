#### **Question:**

Write code for Min-Max Algorithm.

#### **Answer:**

- Agar isMinTurn = True → algorithm compare karta hai left_value aur right_value, aur chhota wala return karta hai (Minimizer ki soch).

- Agar isMinTurn = False → algorithm bara value return karta hai (Maximizer ki soch).

- Har recursive call ek “level” represent karta hai — jaise ek game tree me ek move.

```python
def minimax(node, isMinTurn):

    if type(node) == int:
        return node

    left_value = minimax(node[0], not isMinTurn)
    right_value = minimax(node[1], not isMinTurn)

    if isMinTurn:
        if left_value < right_value:
            return left_value
        else:
            return right_value
    else:
        if left_value > right_value:
            return left_value
        else:
            return right_value


tree = [
    [   # Left branch of root
        [3, 2],   # left subtree
        [5, 4]    # right subtree
    ],
    [   # Right branch of root
        [7, 2],
        [7, 7]
    ]
]

print("Optimal value if minimizer's turn:", minimax(tree, True))
print("Optimal value if maximizer's turn:", minimax(tree, False))
```
