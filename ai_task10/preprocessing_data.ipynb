```python
df['Rating'] = df['Rating'].astype('int64')
df.info()
```

```python
data = pd.DataFrame({'City':df['City'].unique()})
data_LE = LabelEncoder()
data_LE.fit(np.ravel(data))
```

```python
def changeobjtoint(cols):
    for i in cols:
        data = pd.DataFrame({i:df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data))
        # data_LE.fit(data)
        df[i] = data_LE.transform(df[i])
columns = ['Restaurant_Name','Continent','Price_Category','City','Star_Category','Country','Cuisine_Type','Chef','Signature_Dish']
changeobjtoint(columns)
```
