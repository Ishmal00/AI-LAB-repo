### **1. DFS with Stack & Node**


```python
class DepthFirstSearch:
    
    def __init__(self, graph):
        self.graph = graph

    def dfs_stack(self, start, goal):
        visited = []
        stack = [start]

        while stack:
            node = stack.pop()
            if node not in visited:
                visited.append(node)
                if node == goal:
                    return visited
                stack.extend(self.graph[node][::-1])  
        return visited


graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': ['G'],
    'G': []
}

dfs = DepthFirstSearch(graph)
path = dfs.dfs_stack('A', 'F')
print("Visited:", path)
```

```python
class DepthFirstSearch:
    def __init__(self, graph):
        self.graph = graph

    def dfs_stack(self, start, goal):
        visited = []
        stack = [start]

        while stack:
            node = stack.pop()
            if node not in visited:
                print("Visiting node:", node)
                visited.append(node)

                if node == goal:
                    print(f"Goal '{goal}' found! ")
                    return visited


                for child in reversed(self.graph[node]):
                    if child not in visited:
                        stack.append(child)

        print("Goal not found ")
        return visited


graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': ['G'],
    'G': []
}

dfs = DepthFirstSearch(graph)
path = dfs.dfs_stack('A', 'G')

print("\nVisited order:", path)
```

### **2. Research about "Inorder, Preorder, Postorder" and implement in DFS** 

#### ***Tree Traversals (Inorder, Preorder, Postorder)***

Tree traversals thoda alag concept hain. DFS ke hi types hain jo binary trees ke liye define kiye gaye hain. Tumhare case mein tree binary bhi ho sakta hai aur general bhi.

***Preorder (Root → Left → Right)***
Pehle root visit karo, phir left subtree, phir right.
Example: A B D E C F G

POINTS:
- Pehle root node visit hoti hai
- Phir left subtree
- Phir right subtree

***Inorder (Left → Root → Right)***
Pehle left subtree, phir root, phir right.
Example: D B E A F C G
(sirf binary trees mein sense banta hai kyunki "left" aur "right" honay chahiye)

POINTS:
- Pehle left subtree
- Phir root node
- Phir right subtree
Note: Inorder sirf binary tree me sense banata hai (kyunki hamesha ek left aur ek right child hota hai).


***Postorder (Left → Right → Root)***
Pehle left subtree, phir right, phir root.
Example: D E B G F C A

POINTS:
- Pehle left subtree
- Phir right subtree
- Last me root node


#### ***Conclusion :***

Preorder = Root sabse pehle
Inorder = Root beech me
Postorder = Root sabse baad

Ye traversals DFS ke order decide karne ke different rules hain.

```python
tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': ['G'],
    'G': []
}

def preorder(node):
    if node:
        print(node, end=" ")      # Root
        for child in tree[node]:  # Left → Right
            preorder(child)

def inorder(node):
    if node:
        children = tree[node]
        if len(children) >= 1:    # Left child
            inorder(children[0])
        print(node, end=" ")      # Root
        if len(children) == 2:    # Right child
            inorder(children[1])

def postorder(node):
    if node:
        for child in tree[node]:  # Left → Right
            postorder(child)
        print(node, end=" ")      # Root
```

```python
print("Preorder: ")
preorder('A')     

print("\nInorder: ")
inorder('A')      

print("\nPostorder: ")
postorder('A')    
```
